```python
import numpy as np
import matplotlib.pyplot as plt


data1 = np.loadtxt("ET_ASD_10km.txt")
data2 = np.loadtxt("ET_D_asd.txt")
plt.figure(figsize=(8, 4))
# Beide Kurven plotten
plt.loglog(data1[:, 0], data1[:, 1], label='ET DINGO', linewidth=1.5)
plt.loglog(data2[:, 0], data2[:, 1], label='ET mockdata', linewidth=1.5)
# Diagramm-Styling
plt.title('ET ASD Comparison')
plt.xlabel('Frequenz [Hz]')
plt.ylabel('ASD [1/√Hz]')
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.legend()

plt.show()

# Beispielaufruf (Pfade zu deinen .txt Dateien anpassen)
# plot_et_comparison('et_path_1.txt', 'et_path_2.txt')
```
